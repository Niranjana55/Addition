```python
def NBTrain(X_train,y_train,X_test):  
    nb=GaussianNB()
    nb.fit(X_train,y_train)

    prediction=nb.predict(X_test)
    predictionDF=pd.DataFrame(prediction,columns=['Prediction'])
    
    return predictionDF

def DTTrain(X_train,y_train,X_test):  
    dt=DecisionTreeClassifier()
    dt.fit(X_train,y_train)

    prediction=dt.predict(X_test)
    predictionDF=pd.DataFrame(prediction,columns=['Prediction'])
    
    return predictionDF

def LRTrain(X_train,y_train,X_test):  
    lr=LogisticRegression()
    lr.fit(X_train,y_train)

    prediction=lr.predict(X_test)
    predictionDF=pd.DataFrame(prediction,columns=['Prediction'])
    
    return predictionDF

def RFTrain(X_train,y_train,X_test):  
    rf=RandomForestClassifier()
    rf.fit(X_train,y_train)

    prediction=rf.predict(X_test)
    predictionDF=pd.DataFrame(prediction,columns=['Prediction'])
    
    return predictionDF

def GBTTrain(X_train,y_train,X_test):  
    gbt=GradientBoostingClassifier()
    gbt.fit(X_train,y_train)

    prediction=gbt.predict(X_test)
    predictionDF=pd.DataFrame(prediction,columns=['Prediction'])
    
    return predictionDF
```

```python
train=pd.read_csv('KDDTrain.csv')

corr=train.corr()['xAttack']
columnsForUse=corr[corr>0.2]
columnsForUse=list(set(columnsForUse.append(corr[corr<-0.7]).index))

train=train[columnsForUse]

minData=train[train['xAttack']==1]
majData=train[train['xAttack']==0]

algoCount=2

minSize=minData.count()[0]/algoCount
majSize=majData.count()[0]/algoCount

trainData=[]
for _ in range(algoCount):
    trainData.append(pd.concat([minData,majData.sample(majSize)]))

test=pd.read_csv('KDDTest.csv')
test=test[columnsForUse]
X_test=test.drop(['xAttack'],axis=1).values
y_test=test['xAttack'].values
```

```python
X_train=trainData[0].drop(['xAttack'],axis=1).values
y_train=trainData[0]['xAttack'].values
nbPred=GBTTrain(X_train,y_train,X_train)

X_train=trainData[1].drop(['xAttack'],axis=1).values
y_train=trainData[1]['xAttack'].values
dtPred=GBTTrain(X_train,y_train,X_train)

X_train=trainData[0].drop(['xAttack'],axis=1).values
y_train=trainData[0]['xAttack'].values
lrPred=RFTrain(X_train,y_train,X_train)

X_train=trainData[1].drop(['xAttack'],axis=1).values
y_train=trainData[1]['xAttack'].values
rfPred=RFTrain(X_train,y_train,X_train)

nbMetrics=sklearn.metrics.confusion_matrix(nbPred,y_train,labels=[1,0])
dtMetrics=sklearn.metrics.confusion_matrix(dtPred,y_train,labels=[1,0])
lrMetrics=sklearn.metrics.confusion_matrix(lrPred,y_train,labels=[1,0])
rfMetrics=sklearn.metrics.confusion_matrix(lrPred,y_train,labels=[1,0])

nbMetrics=list(nbMetrics.flatten())
nbMetrics.insert(0,'gbt')

dtMetrics=list(dtMetrics.flatten())
dtMetrics.insert(0,'gbt')

lrMetrics=list(lrMetrics.flatten())
lrMetrics.insert(0,'rf')

rfMetrics=list(rfMetrics.flatten())
rfMetrics.insert(0,'rf')

results=pd.DataFrame(columns=['Technique','TP','FP','TN','FN','FPR','TPR','Recall','Precision','TNR','FNR','Accuracy'])
results=results.append(pd.DataFrame([nbMetrics],columns=['Technique','TP','FP','FN','TN']))
results=results.append(pd.DataFrame([dtMetrics],columns=['Technique','TP','FP','FN','TN']))
results=results.append(pd.DataFrame([lrMetrics],columns=['Technique','TP','FP','FN','TN']))
results=results.append(pd.DataFrame([rfMetrics],columns=['Technique','TP','FP','FN','TN']))

results['TPR']=results['TP']/(results['TP']+results['FN'])
results['FPR']=results['FP']/(results['FP']+results['TN'])
results['TNR']=results['TN']/(results['FP']+results['TN'])
results['FNR']=results['FN']/(results['FN']+results['TP'])
results['Recall']=results['TPR']
results['Precision']=results['TP']/(results['TP']+results['FP'])
results['Accuracy']=(results['TP']+results['TN'])/(results['TP']+results['TN']+results['FP']+results['FN'])


bestPositiveAlgo=results[results['TPR']>.8]['Technique'][0]
bestNegativeAlgo=results[results['TNR']>.8]['Technique'][0]
bestAlgo=list(set(results[results['Accuracy']==max(results['Accuracy'])]['Technique'][0]))[0]

resultsPosAgg=pd.DataFrame()

if bestAlgo=='dt':
    Bresults=DTTrain(X_train,y_train,X_test)
elif bestAlgo=='nb':
    Bresults=NBTrain(X_train,y_train,X_test)
elif bestAlgo=='lr':
    Bresults=LRTrain(X_train,y_train,X_test)
elif bestAlgo=='rf':
    Bresults=RFTrain(X_train,y_train,X_test)
elif bestAlgo=='gbt':
    Bresults=GBTTrain(X_train,y_train,X_test)

for a in bestPositiveAlgo:
    if a=='dt':
        results=DTTrain(X_train,y_train,X_test)
    elif a=='nb':
        results=NBTrain(X_train,y_train,X_test)
    elif a=='lr':
        results=LRTrain(X_train,y_train,X_test)
    elif a=='rf':
        results=RFTrain(X_train,y_train,X_test)
    elif a=='gbt':
        results=GBTTrain(X_train,y_train,X_test)
        
    resultsPosAgg=pd.concat([resultsPosAgg,pd.DataFrame(results.values,columns=[a])],axis=1)
    

if len(bestPositiveAlgo)%2==0:
    resultsPosAgg=pd.concat([resultsPosAgg,pd.DataFrame(Bresults.values,columns=[bestAlgo+"Pred"])],axis=1)
    

resultsNegAgg=pd.DataFrame()

for a in bestNegativeAlgo:
    if a=='dt':
        results=DTTrain(X_train,y_train,X_test)
    elif a=='nb':
        results=NBTrain(X_train,y_train,X_test)
    elif a=='lr':
        results=LRTrain(X_train,y_train,X_test)
    elif a=='rf':
        results=RFTrain(X_train,y_train,X_test)
    elif a=='gbt':
        results=GBTTrain(X_train,y_train,X_test)
    
    resultsNegAgg=pd.concat([resultsNegAgg,pd.DataFrame(results.values,columns=[a])],axis=1)
    
if len(bestNegativeAlgo)%2==0:
    resultsNegAgg=pd.concat([resultsNegAgg,pd.DataFrame(Bresults.values,columns=[bestAlgo+"Pred"])],axis=1)


resultsPosAgg['FinalPred']=resultsPosAgg.sum(axis=1)
resultsPosAgg['FinalPred']= resultsPosAgg['FinalPred'].apply(lambda a: 0 if a<1 else 1)

resultsNegAgg['FinalPred']=resultsNegAgg.sum(axis=1)
resultsNegAgg['FinalPred']= resultsNegAgg['FinalPred'].apply(lambda a: 0 if a<=2 else 1)

posResults=resultsPosAgg[resultsPosAgg['FinalPred']==1]
negResults=resultsNegAgg[resultsPosAgg['FinalPred']==0]

pInd=set(posResults.index)
nInd=set(negResults.index)
allInd=set(test.index)
common=pInd.intersection(nInd)
unpredicted=allInd.difference(pInd.union(nInd))
common=common.union(unpredicted)

predTemp=pd.DataFrame(pd.concat([posResults['FinalPred'],negResults['FinalPred']]))

finalPrediction=pd.DataFrame(test['xAttack'])
finalResults=pd.concat([finalPrediction,predTemp],axis=1)

pred=finalResults['FinalPred'].fillna(0.0).astype(int)
actual=finalResults['xAttack'].fillna(0.0).astype(int)

finalConf=sklearn.metrics.confusion_matrix(pred,actual,labels=[1,0])

finalConf=list(finalConf.flatten())
finalConf.insert(0,'final')

final=pd.DataFrame(columns=['Technique','TP','FP','TN','FN','FPR','TPR','Recall','Precision','TNR','FNR','Accuracy'])
final=final.append(pd.DataFrame([finalConf],columns=['Technique','TP','FP','FN','TN']))

final['TPR']=final['TP']/(final['TP']+final['FN'])
final['FPR']=final['FP']/(final['FP']+final['TN'])
final['TNR']=final['TN']/(final['FP']+final['TN'])
final['FNR']=final['FN']/(final['FN']+final['TP'])
final['Recall']=final['TPR']
final['Precision']=final['TP']/(final['TP']+final['FP'])
final['Accuracy']=(final['TP']+final['TN'])/(final['TP']+final['TN']+final['FP']+final['FN'])
final
```
